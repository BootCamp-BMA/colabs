```python
import re

# Function to clean Arabic text
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove mentions and hashtags
    text = re.sub(r"@\S+|#\S+", "", text)

    # Remove special characters and punctuation (except Arabic letters and numbers)
    text = re.sub(r"[^\w\s\u0600-\u06FF]", "", text)

    # Remove Arabic diacritics (Tashkeel)
    text = re.sub(r"[\u064B-\u065F]", "", text)

    # Normalize spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text

# Apply cleaning
df_train["clean_text"] = df_train["text"].apply(clean_text)
df_val["clean_text"] = df_val["text"].apply(clean_text)
df_test["clean_text"] = df_test["text"].apply(clean_text)

# Display sample cleaned text
df_train[["text", "clean_text"]].head()
```

```python
# !pip install jpype1 deep-translator farasapy nltk arabic-reshaper
import random
import nltk
from nltk.corpus import wordnet as wn
from deep_translator import GoogleTranslator
from farasa.segmenter import FarasaSegmenter

# Download Arabic WordNet (if not already downloaded)
nltk.download('omw-1.4')

# Initialize Farasa Segmenter (interactive=False for stability)
segmenter = FarasaSegmenter(interactive=False)

# Function to get Arabic synonyms using WordNet
def get_arabic_synonyms(word):
    synonyms = set()
    for synset in wn.synsets(word, lang='arb'):  # 'arb' is the Arabic language code
        for lemma in synset.lemmas():
            synonyms.add(lemma.name())
    return list(synonyms)

# Function to replace words with synonyms using WordNet
def synonym_replacement(text):
    words = text.split()
    new_words = words.copy()

    for i, word in enumerate(words):
        synonyms = get_arabic_synonyms(word)
        if synonyms:  # If synonyms exist, replace the word
            new_words[i] = random.choice(synonyms)

    return ' '.join(new_words)

# Function for back-translation (Arabic → English → Arabic)
def back_translation(text):
    try:
        translated = GoogleTranslator(source="ar", target="en").translate(text)
        back_translated = GoogleTranslator(source="en", target="ar").translate(translated)
        return back_translated
    except Exception:
        return text  # Return original if translation fails

# Function for random insertion (inserts a synonym)
def random_insertion(text):
    words = text.split()
    if len(words) > 1:
        word = random.choice(words)
        synonyms = get_arabic_synonyms(word)
        if synonyms:
            synonym = random.choice(synonyms)
            words.insert(random.randint(0, len(words)), synonym)
    return ' '.join(words)

# Function for random deletion
def random_deletion(text, p=0.3):
    words = text.split()
    if len(words) == 1:
        return text
    new_words = [word for word in words if random.uniform(0, 1) > p]
    return ' '.join(new_words) if new_words else words[random.randint(0, len(words) - 1)]

# Function to apply all augmentations
def augment_text(text):
    aug1 = synonym_replacement(text)
    aug2 = back_translation(text)
    aug3 = random_insertion(text)
    aug4 = random_deletion(text)
    return [text, aug1, aug2, aug3, aug4]

# Dataset: 10 Arabic sentences
data = [
    "هذا مثال على الجملة الأصلية",
    "معلومة زائفة عن الاقتصاد",
    "خبر مزيف عن الرياضة",
    "التضخم الاقتصادي يزداد هذا العام",
    "حقق المنتخب فوزًا تاريخيًا",
    "انتشرت شائعة حول السياسة",
    "يعتقد البعض أن الأرض مسطحة",
    "تم اختراع تقنية جديدة في الذكاء الاصطناعي",
    "الأسعار ترتفع بشكل غير مسبوق",
    "حقيقة علمية حول الفضاء"
]

# Apply augmentation to all examples
augmented_data = []
for sentence in data:
    augmented_data.append(augment_text(sentence))

# Print results
for i, original in enumerate(data):
    print(f"\nOriginal: {original}")
    print(f"Synonym Replacement: {augmented_data[i][1]}")
    print(f"Back-Translation: {augmented_data[i][2]}")
    print(f"Random Insertion: {augmented_data[i][3]}")
    print(f"Random Deletion: {augmented_data[i][4]}")
```
