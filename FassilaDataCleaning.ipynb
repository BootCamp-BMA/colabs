```python
file_path = '/content/drive/MyDrive/arabic_fake_news/FASSILA/validation_set.csv'
df=pd.read_csv(file_path)
df = df[['text', 'label']]
print(df.shape)
df.head(10)
```

```python
import re



# Remove duplicates
df = df.drop_duplicates()

# Drop rows with null values
df = df.dropna()

def normalize_arabic(text):
    text = str(text)  # Ensure text is a string
    text = re.sub(r'[\u064B-\u065F]', '', text)  # Remove diacritics (Harakat)
    text = text.replace('أ', 'ا').replace('إ', 'ا').replace('آ', 'ا')  # Normalize Alef
    text = text.replace('ى', 'ي').replace('ؤ', 'و').replace('ئ', 'ي')  # Normalize Ya and Waw
    return text

# Apply normalization
df['news'] = df['text'].apply(normalize_arabic)
df = df[['news', 'label']]
# Save cleaned dataset
cleaned_file_path = '/content/drive/MyDrive/arabic_fake_news/FASSILA/cleaned_validation_set.csv'
df.to_csv(cleaned_file_path, index=False)

# Display first few rows
print(df.shape)
df.head()
```

```python

# File paths
train_path = "/content/drive/MyDrive/arabic_fake_news/FASSILA/cleaned_train_set.csv"
test_path = "/content/drive/MyDrive/arabic_fake_news/FASSILA/cleaned_test_set.csv"
validation_path = "/content/drive/MyDrive/arabic_fake_news/FASSILA/cleaned_validation_set.csv"

# Load datasets
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)
df_validation = pd.read_csv(validation_path)

# Combine datasets
df_combined = pd.concat([df_train, df_test, df_validation], ignore_index=True)

# Shuffle the dataset
df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the final combined dataset
final_path = "/content/drive/MyDrive/arabic_fake_news/FASSILA/cleaned_data.csv"
df_combined = df_combined.drop_duplicates()

# Drop rows with null values
df_combined = df_combined.dropna()
print(df_combined.shape)
df_combined.to_csv(final_path, index=False)
# Display first few rows
df_combined.head()
```
